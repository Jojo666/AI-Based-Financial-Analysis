```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from sklearn.metrics import accuracy_score, precision_score, recall_score
import warnings
warnings.filterwarnings('ignore')

# =============================
# 1. Feature Engineering
# =============================
def calculate_features(df, horizon):
    df = df.copy()
    df['Return'] = df['Close'].pct_change()
    df['Risk_Change_5'] = df['RiskScore'].pct_change(5)
    df['PCR_Change_5'] = df['PutToCallRatio'].pct_change(5)
    df['Volume_Change_5'] = df['Volume'].pct_change(5)

    for lag in [1, 2, 3]:
        df[f'PCR_Lag_{lag}'] = df['PutToCallRatio'].shift(lag)
        df[f'Risk_Lag_{lag}'] = df['RiskScore'].shift(lag)
        df[f'Return_Lag_{lag}'] = df['Return'].shift(lag)

    for window in [5, 10]:
        df[f'PCR_Mean_{window}'] = df['PutToCallRatio'].rolling(window).mean()
        df[f'PCR_Std_{window}'] = df['PutToCallRatio'].rolling(window).std()
        df[f'Risk_Mean_{window}'] = df['RiskScore'].rolling(window).mean()
        df[f'Risk_Std_{window}'] = df['RiskScore'].rolling(window).std()
        df[f'Volume_Mean_{window}'] = df['Volume'].rolling(window).mean()
        df[f'Volume_Std_{window}'] = df['Volume'].rolling(window).std()

    df['PCR_Risk_Ratio'] = df['PutToCallRatio'] / df['RiskScore']
    df['PCR_Risk_Product'] = df['PutToCallRatio'] * df['RiskScore']
    df['Return_Volume'] = df['Return'] * df['Volume']
    df['PCR_Return'] = df['PutToCallRatio'] * df['Return']

    df['Target'] = (df['Close'].shift(-horizon) > df['Close']).astype(int)
    return df.dropna()

# =============================
# 2. Model Training
# =============================
def train_model(df, name, horizon, features, high_thresh=0.7, low_thresh=0.3):
    df = df[df['Name'] == name]
    if len(df) < 100:
        return None

    split1 = int(len(df) * 0.6)
    split2 = int(len(df) * 0.8)

    X = df[features].values
    y = df['Target'].astype(int).values

    X_train, X_val, X_test = X[:split1], X[split1:split2], X[split2:]
    y_train, y_val, y_test = y[:split1], y[split1:split2], y[split2:]

    model = xgb.XGBClassifier(
        objective='binary:logistic',
        n_estimators=50,
        max_depth=4,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        use_label_encoder=False,
        eval_metric='logloss',
        n_jobs=-1
    )

    model.fit(X_train, y_train, eval_set=[(X_val, y_val)], verbose=False)

    probas = model.predict_proba(X_test)[:, 1]
    preds = (probas > 0.5).astype(int)

    directions = []
    for p in probas:
        if p >= high_thresh:
            directions.append("Up")
        elif p <= low_thresh:
            directions.append("Down")
        else:
            directions.append("Low Confidence")

    return {
        'Name': name,
        'Horizon': horizon,
        'Accuracy': accuracy_score(y_test, preds),
        'Up Precision': precision_score(y_test, preds, zero_division=0),
        'Down Precision': precision_score(y_test, preds, pos_label=0, zero_division=0),
        'Up Recall': recall_score(y_test, preds, pos_label=1, zero_division=0),
        'Down Recall': recall_score(y_test, preds, pos_label=0, zero_division=0),
        'Feature Importance': dict(zip(features, model.feature_importances_)),
        'Avg Confidence': np.mean(np.maximum(probas, 1 - probas)),
        'Predicted Direction': directions[-1],
        'Confidence': probas[-1]
    }

# =============================
# 3. Run Full Pipeline
# =============================
def run_pipeline_with_confidence(df_with_pcr):
    tickers = df_with_pcr['Name'].unique()
    horizons = [1, 3, 5, 10, 15, 20]
    all_results = []

    for horizon in horizons:
        print(f"Processing {horizon}-day horizon...")
        df_all = pd.concat([
            calculate_features(df_with_pcr[df_with_pcr['Name'] == t], horizon)
            for t in tickers
        ])

        drop_cols = ['Date', 'Name', 'Target', 'Close']
        features = [col for col in df_all.select_dtypes(include=np.number).columns if col not in drop_cols]

        for ticker in tickers:
            result = train_model(df_all, ticker, horizon, features)
            if result:
                all_results.append(result)

    return pd.DataFrame(all_results)
```

```python
results_df = run_pipeline_with_confidence(df_with_pcr)
```

```python
# =============================
# 4. Display Results and Forward Table
# =============================
def display_confidence_summary(results_df, horizon_focus=20):
    display_df = results_df[results_df['Horizon'] == horizon_focus].copy()

    display_df = display_df[['Name', 'Horizon', 'Predicted Direction', 'Confidence']]
    display_df['Confidence'] = (display_df['Confidence'] * 100).round(2).astype(str) + '%'
    display_df = display_df.rename(columns={
        'Name': 'Asset',
        'Horizon': 'Time Window',
        'Predicted Direction': 'Predicted Direction***'
    })

    print("\nChart 1: Forward looking directional table")
    print(display_df.to_string(index=False))

    # Portfolio summary
    direction_counts = display_df['Predicted Direction***'].value_counts()
    total = len(display_df)
    print("\nChart 2: Portfolio-Level Prediction Summary")
    print("Direction\tCount\t% of Total")
    for direction in ['Up', 'Down', 'Low Confidence']:
        count = direction_counts.get(direction, 0)
        pct = round(100 * count / total)
        print(f"{direction}\t\t{count}\t{pct}%")

# Run this after running your pipeline
display_confidence_summary(results_df, horizon_focus=20)  # or change to 5, 10 etc.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from sklearn.metrics import accuracy_score, precision_score, recall_score
import warnings
warnings.filterwarnings('ignore')

# =============================
# 1. Feature Engineering
# =============================
def calculate_features(df, horizon):
    df = df.copy()

    # Core signals
    df['Return'] = df['Close'].pct_change()
    df['Risk_Change_5'] = df['RiskScore'].pct_change(5)
    df['PCR_Change_5'] = df['PutToCallRatio'].pct_change(5)
    df['Volume_Change_5'] = df['Volume'].pct_change(5)

    # Lag features
    for lag in [1, 2, 3]:
        df[f'PCR_Lag_{lag}'] = df['PutToCallRatio'].shift(lag)
        df[f'Risk_Lag_{lag}'] = df['RiskScore'].shift(lag)
        df[f'Return_Lag_{lag}'] = df['Return'].shift(lag)

    # Rolling statistics
    for window in [5, 10]:
        df[f'PCR_Mean_{window}'] = df['PutToCallRatio'].rolling(window).mean()
        df[f'PCR_Std_{window}'] = df['PutToCallRatio'].rolling(window).std()
        df[f'Risk_Mean_{window}'] = df['RiskScore'].rolling(window).mean()
        df[f'Risk_Std_{window}'] = df['RiskScore'].rolling(window).std()
        df[f'Volume_Mean_{window}'] = df['Volume'].rolling(window).mean()
        df[f'Volume_Std_{window}'] = df['Volume'].rolling(window).std()

    # Interaction features
    df['PCR_Risk_Ratio'] = df['PutToCallRatio'] / df['RiskScore']
    df['PCR_Risk_Product'] = df['PutToCallRatio'] * df['RiskScore']
    df['Return_Volume'] = df['Return'] * df['Volume']
    df['PCR_Return'] = df['PutToCallRatio'] * df['Return']

    # Target variable
    df['Target'] = (df['Close'].shift(-horizon) > df['Close']).astype(int)
    return df.dropna()

# =============================
# 2. Model Training
# =============================
def train_model(df, name, horizon, features):
    df = df[df['Name'] == name]
    if len(df) < 100:
        return None

    split1 = int(len(df) * 0.6)
    split2 = int(len(df) * 0.8)

    X = df[features].values
    y = df['Target'].astype(int).values

    X_train, X_val, X_test = X[:split1], X[split1:split2], X[split2:]
    y_train, y_val, y_test = y[:split1], y[split1:split2], y[split2:]

    model = xgb.XGBClassifier(
        objective='binary:logistic',
        n_estimators=50,
        max_depth=4,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        use_label_encoder=False,
        eval_metric='logloss',
        n_jobs=-1
    )

    model.fit(X_train, y_train, eval_set=[(X_val, y_val)], verbose=False)

    preds = model.predict(X_test)
    return {
        'Name': name,
        'Horizon': horizon,
        'Accuracy': accuracy_score(y_test, preds),
        'Up Precision': precision_score(y_test, preds, zero_division=0),
        'Down Precision': precision_score(y_test, preds, pos_label=0, zero_division=0),
        'Up Recall': recall_score(y_test, preds, pos_label=1, zero_division=0),
        'Down Recall': recall_score(y_test, preds, pos_label=0, zero_division=0),
        'Feature Importance': dict(zip(features, model.feature_importances_))
    }

# =============================
# 3. Main Pipeline
# =============================
def run_pipeline_all_features(df_with_pcr):
    tickers = df_with_pcr['Name'].unique()
    horizons = [1, 3, 5, 10, 15, 20]
    all_results = []

    for horizon in horizons:
        print(f"Processing {horizon}-day horizon...")
        df_all = pd.concat([
            calculate_features(df_with_pcr[df_with_pcr['Name'] == t], horizon)
            for t in tickers
        ])

        drop_cols = ['Date', 'Name', 'Target', 'Close']
        features = [col for col in df_all.select_dtypes(include=np.number).columns if col not in drop_cols]

        for ticker in tickers:
            result = train_model(df_all, ticker, horizon, features)
            if result:
                all_results.append(result)

    return pd.DataFrame(all_results)

# =============================
# 4. Print Summary Table
# =============================
def print_summary(results_df):
    print("\n==== PERFORMANCE SUMMARY ====\n")
    for horizon in sorted(results_df['Horizon'].unique()):
        print(f"HORIZON: {horizon} days")
        subset = results_df[results_df['Horizon'] == horizon]
        print(subset[['Name', 'Accuracy', 'Up Precision', 'Down Precision']].sort_values(by='Accuracy', ascending=False))
        print("-" * 60)

# =============================
# 5. Visualizations
# =============================
def visualize_results(results_df):
    for metric in ['Accuracy', 'Up Precision', 'Down Precision']:
        plt.figure(figsize=(12, 6))
        sns.barplot(data=results_df, x='Horizon', y=metric, hue='Name')
        plt.axhline(0.5, color='red', linestyle='--', label='Random Baseline')
        plt.title(f'{metric} Across Horizons')
        plt.legend()
        plt.tight_layout()
        plt.show()

    for horizon in sorted(results_df['Horizon'].unique()):
        subset = results_df[results_df['Horizon'] == horizon]
        for _, row in subset.iterrows():
            feat_df = pd.DataFrame(list(row['Feature Importance'].items()), columns=['Feature', 'Importance'])
            feat_df = feat_df.sort_values(by='Importance', ascending=False)
            plt.figure(figsize=(8, 4))
            sns.barplot(x='Importance', y='Feature', data=feat_df, palette='coolwarm')
            plt.title(f"{row['Name']} - Feature Importance (Horizon = {horizon} days)")
            plt.tight_layout()
            plt.show()

    heatmap_data = results_df.pivot(index='Name', columns='Horizon', values='Accuracy')
    plt.figure(figsize=(10, 6))
    sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap='YlGnBu')
    plt.title("Accuracy Heatmap by Stock and Horizon")
    plt.tight_layout()
    plt.show()
```

```python
results_df = run_pipeline_all_features(df_with_pcr)
print_summary(results_df)
visualize_results(results_df)
```
